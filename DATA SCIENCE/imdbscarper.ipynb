```python
 from bs4 import BeautifulSoup # for data reading purpose, this is library
import requests
```

```python
data = requests.get("https://www.imdb.com/chart/top/")
```

```python
if data.status_code ==200:
    print("success 200")
elif data.status_code ==404:
    print("page not found")
elif data.status_code ==500:
    print("server error")    
        
```

```python
soup = BeautifulSoup(data.text)
```

```python
rows = soup.find('tbody').findAll('tr')
```

```python
len(rows)
```

```python
for movie in rows:
    print(movie.find('td',{'class' : 'titleColumn'}).find('a').text)
    
```

```python
for movie in rows:
    print(movie.find('td',{'class' : 'titleColumn'}).find('a').text)
    print(movie.find('td',{'class' : 'ratingColumn imdbRating'}).text)
    print(movie.find('span',{'class' : 'secondaryInfo'}).text)
    
```

```python
movie_detail = []

for movie in rows:
  detail = {}
  detail['Name'] = movie.find('td',{'class' : 'titleColumn'}).find('a').text
  detail['Year'] = movie.find('td',{'class' : 'titleColumn'}).find('span').text.strip('()')
  detail['Rating'] = movie.find('span',{'class' : 'secondaryInfo'}).text

  movie_detail.append(detail)  
```

```python
movie_detail
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(movie_detail)
```
