```python
df['Total']=df.sum(axis=1)
```

```python
df
```

```python
df.groupby('AreaName').sum()
```

```python
AreaName= df.groupby('AreaName').sum()['Total']
```

```python
AreaName.plot(kind="pie",autopct="%1.2f%%", figsize=(20,10))
```

```python
explode = [0,0,0.2,0,0.3,0]
AreaName.plot(kind="pie",explode=explode, shadow=True, autopct="%1.2f%%", figsize=(20,10),labels= None)
plt.legend(AreaName.index, loc='upper left')
```

```python
AreaName.plot(kind="pie",autopct="%1.2f%%", pctdistance=0.7, figsize=(20,10),labels= None, wedgeprops={ 'width': 0.5})
plt.legend(AreaName.index, loc='upper left')
```
