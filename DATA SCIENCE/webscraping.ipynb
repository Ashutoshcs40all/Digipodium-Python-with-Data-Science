```python
url1 = "https://timesofindia.indiatimes.com/news"
```

```python
data = requests.get(url1)
```

```python
data.status_code
```

```python
soup1 = BeautifulSoup(data.text)
```

```python
soup1.find("h1").text
```

```python
rows = soup1.findAll('li', { 'class' : 'prime' })
```

```python
len(rows)
```

```python
for news in soup1.find('div', {'class' : 'main content'}).findAll('li',{'class'} : 'prime' }):
    print(news.find('span').find('a').text)
    summary = news.find('span').find('w_desc')

    if summary:
        print(summary.text)
    else:
        print('No summary')


    #print('===================
    #======================================
    #==========================')       
```

```python
for news in rows:
    print(news.find('h2').text)
    print(news.find('p').text)
    print(news.find('img').attrs.get('src'))
```
